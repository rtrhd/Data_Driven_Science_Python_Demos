```python
# Python control toolbox: https://python-control.readthedocs.io/
from control import tf, bode, use_numpy_matrix
use_numpy_matrix(False)

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 16
```

```python
# Figure 8.27

s = tf([1, 0], [0, 1])
L = 1/s
S = 1/(1 + L)
T = L/(1 + L)

plt.figure(figsize=(10, 10))
bode(L, label='L', color='C0', dB=True)
bode(S, label='S', color='C3', dB=True)
bode(T, label='T', color='C1', dB=True)
plt.legend(fontsize='small')
plt.suptitle('Bode Diagram')
plt.tight_layout();
```
