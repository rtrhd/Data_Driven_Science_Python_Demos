```python
import numpy as np
from math import pi as π
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
t = np.arange(0, 10, 0.01)      # Time

w_r = np.full_like(t, 60)       # Constant reference speed
w_d = 10 * np.sin(π * t)        # Sinusoidal disturbance at 0.5Hz

a_model = 1                     # Model system: y = a_model * u
a_true = 0.5                    # True system: y = a_true * u

u_open = w_r / a_model          # Open-loop u based on model
y_open = a_true * u_open + w_d  # Open-loop response


# Control law u = K(w_r - y), with gain K
K = 50
y_closed = a_true * K / (1 + a_true*K) * w_r + w_d / (1 + a_true*K)
```

```python
# Figure 8.5

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(t, w_r, color='black', linewidth=2, label='Reference')
ax.plot(t, w_d, '--', color='black', linewidth=1.5, label='Disturbance')
ax.plot(t, y_open, color='red', linewidth=1.5, label='Open Loop')
ax.plot(t, y_closed, color='blue', linewidth=1.5, label='Closed Loop')

ax.set_xlabel('Time')
ax.set_ylabel('Speed')
ax.legend(loc='lower right', fontsize='small', framealpha=1.0)

ax.grid(True)
```
