```python
import numpy as np
from numpy.linalg import svd
from skimage.transform import rotate

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
n = 1000
q = n // 4
X = np.zeros((n, n))
X[q:3*q, q:3*q] = 1

X_rot = (rotate(X, 10) > 0).astype(int)

fig, axs = plt.subplots(1, 2, figsize=(16, 8))
axs[0].imshow(X, cmap='gray')
axs[0].axis(False)
axs[0].set_title('Unrotated Matrix')

axs[1].imshow(X_rot, cmap='gray')
axs[1].axis(False)
axs[1].set_title('Rotated Matrix');
```

```python
U, S, VT = svd(X, full_matrices=False)
U_rot, S_rot, VT_rot = svd(X_rot, full_matrices=False)

fig, axs = plt.subplots(1, 2, sharey=True, figsize=(16, 8))

axs[0].semilogy(S, '-o', mfc='none')
axs[0].grid(True)
axs[0].set_title('Unrotated Matrix: Spectrum');

axs[1].semilogy(S_rot, '-o',  mfc='none')
axs[1].grid(True)
axs[1].set_title('Rotated Matrix: Spectrum');
```
