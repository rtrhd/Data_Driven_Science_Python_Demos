```python
import numpy as np
from numpy.random import randn, rand

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
plt.rcParams['font.size'] = 18
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
x = np.arange(0.2, 5, 0.1)
n = len(x)
y = 1 / x

y2 = y + 0.5*randn(n)
y3s = y2[np.newaxis, :] + 2*rand(5, n) + 1

ax.plot(x, y, color='black', linewidth=2)
ax.add_patch(
    Rectangle(
        (0.5, 0.4), 0.9, 1.6,
        linewidth=1, edgecolor='black', facecolor='grey', alpha=0.5
    ) 
)

ax.scatter(x, y2, 100, color='darkmagenta', edgecolors='black')
for y3 in y3s:
    ax.scatter(x, y3, 100, color='green', edgecolors='black')

ax.set_xlim(0.2, 4)
ax.set_ylim(0, 5.5)
ax.grid(True)
ax.set_xlabel('Number of Terms')
ax.set_ylabel('Error')
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.text(0.6, 1.0, 'Pareto\nOptimal')
ax.text(2.5, 4.0, 'Models')
ax.text(2.5, 1.0, 'Pareto Frontier');
```
