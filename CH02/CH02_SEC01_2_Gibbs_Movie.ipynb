```python
import numpy as np
from numpy import pi as Ï€

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.rcParams['font.size'] = 18

from IPython.display import HTML, display
```

```python
# Figure 2.5

dx = 0.01
L = 10
x = np.arange(0, L+dx, dx)
n = len(x)
nquart = n // 4

f = np.zeros_like(x)
f[nquart: 3*nquart] = 1

A0 = f.sum() * dx * 2 / L
fFS = np.full_like(f, A0/2)

all_fFS = np.zeros((len(fFS), 101))
all_fFS[:, 0] = fFS
for k in range(1, 101):
    Ak = np.sum(f * np.cos(2*Ï€ * k*x / L)) * dx * 2 / L
    Bk = np.sum(f * np.sin(2*Ï€ * k*x / L)) * dx * 2 / L
    fFS = fFS + Ak * np.cos(2*Ï€ * k*x / L) + Bk * np.sin(2*Ï€ * k*x / L)
    all_fFS[:, k] = fFS

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, f, linewidth=2)
fFS_plot, = ax.plot([], [], linewidth=2)
ax.grid(True)

def init():
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(-0.2, 1.2)

def animate(iter_):
    fFS_plot.set_data(x, all_fFS[:, iter_])

anim = FuncAnimation(fig, animate, init_func=init, frames=101, interval=50, blit=False, repeat=False)
display(HTML(anim.to_jshtml()))
plt.close()
```
