```python
import numpy as np
from scipy.interpolate import interp1d

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
f = np.array(
    [0, 0, 0.1, 0.2, 0.25, 0.2, 0.25, 0.3, 0.35, 0.43, 0.45, 0.5, 0.55, 0.5, 0.4, 0.425, 0.45, 0.425, 0.4, 0.35, 0.3, 0.25, 0.225, 0.2, 0.1, 0, 0]
)
g = np.array(
    [0, 0, 0.025, 0.1, 0.2, 0.175, 0.2, 0.25, 0.25, 0.3, 0.32, 0.35, 0.375, 0.325, 0.3, 0.275, 0.275, 0.25, 0.225, 0.225, .2, 0.175, 0.15, 0.15, 0.05, 0, 0]
) - 0.025

x = 0.1 * np.arange(1, len(f) + 1)
xf = np.arange(0.1, x[-1], 0.01)

f_interp = interp1d(x, f, kind='cubic')
g_interp = interp1d(x, g, kind='cubic')

ff = f_interp(xf)  
gf = g_interp(xf)
```

```python
# Figure 2.1

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(xf[10:-10], ff[10:-10], color='black', linewidth=2)
ax.plot(x[1:-2], f[1:-2], 'o')
ax.plot(xf[10:-10], gf[10:-10], color='black', linewidth=2)
ax.plot(x[1:-2], g[1:-2], 'o')
ax.grid(True);
```
