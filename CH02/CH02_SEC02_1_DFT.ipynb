```python
import numpy as np
from numpy import pi as Ï€

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
n = 256
Ï‰ = np.exp(-1j * 2*Ï€ / n)

# Slow
dft_python = np.zeros((n, n), dtype=complex)
for i in range(n):
    for k in range(n):
        dft_python[i, k] = Ï‰ ** (i * k)

# Fast
J, K = np.meshgrid(np.arange(n), np.arange(n))
dft_numpy = Ï‰ ** (J * K)

assert np.allclose(dft_python, dft_numpy)

fig, ax = plt.subplots(1, 1, figsize=(12, 12))
im = ax.imshow(dft_numpy.real, cmap='turbo')
ax.axis(False)
fig.colorbar(
    im,
    cax=make_axes_locatable(ax).append_axes('right', size='5%', pad=0.2)
);
```
