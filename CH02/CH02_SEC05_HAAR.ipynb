```python
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
x = np.arange(0, 1, 0.001)
n = len(x)
n2 = n // 2
n4 = n // 4

f10 = np.zeros_like(x)
f10[:n2] = 1
f10[n2:] = -1

f21 = np.zeros_like(x)
f21[:n4] = 1
f21[n4:n2] = -1
f21 = f21 * np.sqrt(2)

f22 = np.zeros_like(x)
f22[n2: (n2 + n4)] = 1
f22[(n2 + n4):] = -1
f22 = f22 * np.sqrt(2)

x = np.append([-1, 0], x)
x = np.append(x, [1, 2])

f10 = np.pad(f10, (2, 2), 'constant')
f21 = np.pad(f21, (2, 2), 'constant')
f22 = np.pad(f22, (2, 2), 'constant')

fig, axs = plt.subplots(3, 1, figsize=(12, 8))
axs[0].plot(x, f10, linewidth=2)
axs[0].set_xlim(-0.2, 1.2)
axs[0].set_ylim(-1.75, 1.75)
axs[0].grid(True)
axs[1].plot(x, f21, linewidth=2)
axs[1].set_xlim(-0.2, 1.2)
axs[1].set_ylim(-1.75, 1.75)
axs[1].grid(True)
axs[2].plot(x, f22, linewidth=2)
axs[2].set_xlim(-0.2, 1.2)
axs[2].set_ylim(-1.75, 1.75)
axs[2].grid(True);
```

```python
x = np.arange(-5, 5, 0.001)
f_mex_hat = (1 - x**2) * np.exp(-x**2 / 2)
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, f_mex_hat, linewidth=2)
ax.grid(True);
```
