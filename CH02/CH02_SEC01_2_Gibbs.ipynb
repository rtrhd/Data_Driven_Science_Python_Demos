```python
import numpy as np
from numpy import pi as π

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
# Figure 1.25

dx = 0.01
L = 2*π
x = np.arange(0, L+dx, dx)
n = len(x)
nquart = n // 4

f = np.zeros_like(x)
f[nquart: 3*nquart] = 1

A0 = f.sum() * dx * 2 / L
fFS = np.full_like(f, A0/2)

for k in range(1,101):
    Ak = np.sum(f * np.cos(2*π * k*x / L)) * dx * 2 / L
    Bk = np.sum(f * np.sin(2*π * k*x / L)) * dx * 2 / L
    fFS = fFS + Ak * np.cos(2*π * k*x / L) + Bk * np.sin(2*π * k*x / L)

fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, f, color='black', linewidth=2)
ax.plot(x, fFS, '-', color='red', linewidth=1.5)
ax.grid(True);
```
