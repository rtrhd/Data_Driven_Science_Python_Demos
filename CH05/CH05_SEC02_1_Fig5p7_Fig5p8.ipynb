```python
import numpy as np
from numpy import pi as π
from numpy.random import randn

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
# Training and test set sizes
n1 = 100 # Training set size
n2 = 50  # Test set size
n = n1 + n2

# Random ellipse 1 centered at (-2, 0)
x, y = randn(n) - 2, 0.5*randn(n)

# Random ellipse 5 centered at (1, 0)
x5, y5 = 2*randn(n) + 1, 0.5*randn(n)

# Random ellipse 2 centered at (2, -2)
x2, y2 = randn(n) + 2, 0.2*randn(n) - 2

# Rotate ellipse 2 by angle θ
θ = π/4
A = np.array([
    [np.cos(θ), -np.sin(θ)],
    [np.sin(θ),  np.cos(θ)]
])

x3 = A[0, 0] * x2 + A[0, 1] * y2
y3 = A[1, 0] * x2 + A[1, 1] * y2
```

```python
# Figure 5.7

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(12, 10))

axs[0][0].plot(x[:n1], y[:n1], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[0][0].plot(x3[:n1], y3[:n1], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[0][0].set_title('Unsupervised')

axs[0][1].plot(x[:70], y[:70], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[0][1].plot(x3[:70], y3[:70], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[0][1].plot(x[70:100], y[70:100],'o', markerfacecolor='green', markeredgecolor='black', ms=10)
axs[0][1].plot(x3[70:100], y3[70:100],'o', markerfacecolor='darkmagenta', markeredgecolor='black', ms=10)
axs[0][1].set_title('Supervised')

axs[1][0].plot(x5[:n1], y5[:n1], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[1][0].plot(x3[:n1], y3[:n1], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)

axs[1][1].plot(x5[:70], y5[:70], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[1][1].plot(x3[:70], y3[:70], 'o', markerfacecolor='darkgray', markeredgecolor='black', ms=10)
axs[1][1].plot(x5[70:100], y5[70:100], 'o', markerfacecolor='green', markeredgecolor='black', ms=10)
axs[1][1].plot(x3[70:100], y3[70:100], 'o', markerfacecolor='darkmagenta', markeredgecolor='black', ms=10)

for ax in axs.flat:
    ax.set_xlim(-6, 6)
    ax.set_ylim(-2, 2)
    ax.set_xticks([-5, 0, 5])
    ax.set_yticks([-2, -1, 0, 1, 2])

plt.tight_layout()
```

```python
# Figure 5.8

n1 = 300 # training set size
x1 = 1.5*randn(n1) - 1.5
y1 = 1.2*randn(n1) + (x1 + 1.5)**2 - 7
x2 = 1.5*randn(n1) + 1.5
y2 = 1.2*randn(n1) - (x2 - 1.5)**2 + 7

fig, axs = plt.subplots(1, 2, figsize=(13, 5))

axs[0].plot(x1, y1, 'o', markerfacecolor='green', markeredgecolor='black', ms=10)
axs[0].plot(x2, y2, 'o', markerfacecolor='darkmagenta', markeredgecolor='black', ms=10)
axs[0].set_xlim(-6, 6)
axs[0].set_ylim(-12, 12)
axs[0].set_xticks([-5, 0, 5])
axs[0].set_yticks([-10, -5, 0, 5, 10])

r = 7 + randn(n1)
θ = 2*π * randn(n1)
xr = r * np.cos(θ)
yr = r * np.sin(θ)

x5 = randn(n1)
y5 = randn(n1)

axs[1].plot(xr, yr, 'o', markerfacecolor='green', markeredgecolor='black', ms=10)
axs[1].plot(x5, y5, 'o', markerfacecolor='darkmagenta', markeredgecolor='black', ms=10)
axs[1].set_xlim(-10, 10)
axs[1].set_ylim(-10, 10)
axs[1].set_xticks([-10, -5, 0, 5, 10])
axs[1].set_yticks([-10, -5, 0, 5, 10])

plt.tight_layout()
```
